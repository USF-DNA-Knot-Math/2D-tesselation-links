```python
def get_long_example(height=2):
    height=height-1
    start=height+1
    ups=[[('b',start+2*i), ('B', start+2+2*i)] for i in range(height)]
    loop1=[[('b',start+2*height), ('b',start+1+2*height)]]
    downs=[[('B', start+1+2*i),('b', start+1+2*(i-1))] for i in range(height,0,-1)]
    loop2=[[('B', start+1), ('B',start+2+2*height)], [('b',start+2+2*height),('B',1)]]
    closing=[[('b',i+1),('B',i+2)] for i in range(height)]
    Chords_raw=pretty_print_Chords(ups+loop1+downs+loop2+closing,show_print=False)
    return Chords_raw
```
